**The dataset is from an ongoing cardiovascular study on residents of the town of Framingham,**
**Massachusetts. The classification goal is to predict whether the patient has a 10-year risk of**
**future coronary heart disease (CHD). The dataset provides the patients information. It includes**
**over 4,000 records and 15 attributes.**
**Variables**
**Each attribute is a potential risk factor. There are both demographic, behavioral, and medical risk**
**factors.**

**Data Description**
**Demographic:**

** Sex: male or female("M" or "F")**

** Age: Age of the patient;(Continuous - Although the recorded ages have been truncated to**
**whole numbers, the concept of age is continuous)**

**Behavioral**

** is_smoking: whether or not the patient is a current smoker ("YES" or "NO")**

** Cigs Per Day: the number of cigarettes that the person smoked on average in one day.(can be**
**considered continuous as one can have any number of cigarettes, even half a cigarette.)**

**Medical( history)**

** BP Meds: whether or not the patient was on blood pressure medication (Nominal)**

** Prevalent Stroke: whether or not the patient had previously had a stroke (Nominal)**

** Prevalent Hyp: whether or not the patient was hypertensive (Nominal)**

** Diabetes: whether or not the patient had diabetes (Nominal)**

**Medical(current)**

** Tot Chol: total cholesterol level (Continuous)**

** Sys BP: systolic blood pressure (Continuous)**

** Dia BP: diastolic blood pressure (Continuous)**

** BMI: Body Mass Index (Continuous)**

** Heart Rate: heart rate (Continuous - In medical research, variables such as heart rate though in**
**fact discrete, yet are considered continuous because of large number of possible values.)**

** Glucose: glucose level (Continuous)**

**Predict variable (desired target)**

** 10-year risk of coronary heart disease CHD(binary: 1, means Yes, 0** **means No)**

# **Handling imbalance data using SMOTE**

####**WHAT IS SMOTE ?**
This is a statistical technique for increasing the number of cases in your dataset in a balanced way. The module works by generating new instances from existing minority cases that you supply as input. This implementation of SMOTE does not change the number of majority cases. SMOTE takes the entire dataset as an input, but it increases the percentage of only the minority cases.

##Specificity (True Negative Rate) is calculated as <code>TN / (TN + FP)</code>, where TN is the number of true negatives, and FP is the number of false positives. It measures the model's ability to correctly identify the negative class.

##Sensitivity (True Positive Rate) is calculated as  <code>TP / (TP + FN) </code>, where TP is the number of true positives, and FN is the number of false negatives. It measures the model's ability to correctly identify the positive class.

##These metrics provide insights into how well the model performs in distinguishing between positive and negative cases. High sensitivity indicates the model's ability to catch most positive cases, while high specificity indicates the model's ability to exclude most negative cases. The choice of which metric to prioritize depends on the specific context and the importance of correctly identifying one class over the other.

```python
# Specificity (True Negative Rate)
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp)

# Sensitivity (True Positive Rate)
sensitivity = tp / (tp + fn)
```

# **8.Support Vector Classifier**

```python
# importing support vector classifier
from sklearn.svm import SVC
```

The param_grid you've defined is used for tuning hyperparameters of a Support Vector Machine (SVM) classifier. SVM is a powerful algorithm for classification and regression tasks. Let's break down the hyperparameters and their values in the param_grid for SVM:

C: This hyperparameter controls the trade-off between maximizing the margin and minimizing the classification error. Smaller values of C result in a larger margin but may allow some misclassifications, while larger values of C minimize misclassifications but may result in a narrower margin. You've specified three values for C: 0.1, 1, and 10.

gamma: This hyperparameter controls the shape of the decision boundary. Smaller values of gamma lead to a more flexible decision boundary, while larger values make it more rigid. You've specified three values for gamma: 1, 0.1, and 0.01.

kernel: The kernel function is used to transform the input features into a higher-dimensional space where the classes are more separable. Different kernel functions have different characteristics. You've specified three kernel functions: 'linear', 'rbf' (Radial Basis Function), and 'sigmoid'.

probability: Setting this to True allows the SVM classifier to estimate class probabilities. It's useful when you want to assess not just the predicted class labels but also the probability associated with each class.

With this param_grid, you're essentially instructing scikit-learn to perform a grid search over these hyperparameters and their specified values to find the combination that results in the best model performance. GridSearchCV or RandomizedSearchCV from scikit-learn can be used with this param_grid to find the optimal hyperparameter values for your SVM classifier.

The search will train multiple SVM models with different combinations of C, gamma, kernel, and probability, and evaluate their performance, typically using cross-validation. After the search is complete, you can access the best combination of hyperparameters to train your final SVM model with the most suitable settings for your specific problem.

```python
#Support vector classifier
svc = SVC()

#running grid search for optimal parameter
param_grid = {
    'C': [0.1,1,10],
    'gamma':[1,0.1,0.01],
    'kernel':['linear','rbf','sigmoid'],
    'probability':[True]
}

svc_classifier = GridSearchCV(svc,param_grid,cv=2)

#fitting the model
svc_classifier.fit(X_train,y_train)

#Predicting
y_pred_svc = svc_classifier.predict(X_test)

#predicting probability
y_pred_proba_svc = svc_classifier.predict_proba(X_test)[:,1]
```
